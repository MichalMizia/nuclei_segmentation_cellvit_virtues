```python
from matplotlib.patches import Patch
cell_type_labels = [
    None, "Stroma", "Endothelia", "Macrophage", "Tumor",
    "Helper T cell", "Cytotoxic T cell", "Other", "Treg", "B cells"
]

def visualize_images(ds, num_cols=1):
    he_ds = ds.unimodal_datasets['he']
    tissue_ids = ds.get_tissue_ids()
    cmap = plt.get_cmap('tab10')

    plt.figure(figsize=(num_cols * 5, 2 * 5))
    plt.suptitle("Tissue Images")
    for i in range(num_cols):
        img = rearrange(he_ds._get_tissue_all_channels(tissue_ids[i]), 'c h w -> h w c')
        img = (img - img.min()) / (img.max() - img.min())
        mask = ds.get_cell_mask(tissue_ids[i], task="broad_cell_type", resize=True)
        
        plt.subplot(2, num_cols, i + 1)
        plt.imshow(img)
        plt.axis('off')
        plt.title(f"Tissue {i+1}")

        # Second row: image + mask
        plt.subplot(2, num_cols, num_cols + i + 1)
        plt.imshow(img)
        plt.imshow(np.ma.masked_where(mask == 0, mask), cmap=cmap, alpha=0.9, vmin=0, vmax=9)
        plt.axis('off')
        plt.title(f"Tissue {i+1} + Mask")

        if i == 0:
            unique_vals = np.unique(mask)
            legend_elements = []
            for val in unique_vals:
                if val == 0 or cell_type_labels[val] is None:
                    continue
                legend_elements.append(
                    Patch(facecolor=cmap(val), edgecolor='k', label=cell_type_labels[val])
                )
            if legend_elements:
                plt.legend(handles=legend_elements)
```

## CellVit Structure

The cellvit pypi documentation provides very little information about the usage of python package. Instead it only shows how to use the CLI. Here we get to know the structure of the CellVit package and the list of its modules

```python
import cellvit
cellvit_dir = os.path.dirname(cellvit.__file__)
print(cellvit_dir)
sys.path.insert(0, cellvit_dir)

modules = []
for root, dirs, files in os.walk(cellvit_dir):
    for file in files:
        if file.endswith(".py") and file != "__init__.py":
            rel_path = os.path.relpath(os.path.join(root, file), cellvit_dir)
            module_name = rel_path.replace(os.path.sep, ".").replace(".py", "")
            modules.append(module_name)

print("Modules in cellvit package:")
print(modules)
```

## You can download the model from https://figshare.com/ndownloader/files/45351934

```python
from cellvit.models.cell_segmentation.cellvit_sam import CellViTSAM

weights_path = "/home/mizia/.cache/cellvit/CellViT-SAM-H-x40-AMP.pth" # replace with your path
device = "cuda"

model = CellViTSAM(
    model_path=weights_path,
    vit_structure="SAM-H",
    num_nuclei_classes=10,
    num_tissue_classes=19, # only used for tissue classification, we can set it to anything
)
model.to(device)
model.eval()
```

```python
x = torch.rand(8, 3, 256, 256).to(device)
x *= 255
with torch.no_grad():
    with torch.autocast(device_type="cuda", dtype=torch.float16):
        out = model(x, retrieve_tokens=True)
pred = out["nuclei_type_map"].argmax(dim=1)
print(pred.shape)
print("unique predicted classes:", torch.unique(pred))
```

```python
from tiler import Tiler, Merger

img = get_image(tissue_ids[0]).numpy()
C, H, W = img.shape

patch_size = 256
overlap = 64 # overlap to avoid edge artifacts
batch_size = 8

tiler = Tiler(
    data_shape=img.shape,
    tile_shape=(C, patch_size, patch_size),
    overlap=overlap,
    channel_dimension=0
)
merger = Merger(tiler)

for batch_id, batch in tiler(img, batch_size=batch_size):
    merger.add_batch(batch_id, batch_size, batch) # batch is np (B, C, P, P)
    # possibly
    # tile_torch = torch.from_numpy(tile).permute(2, 0, 1).float().to(device)
    # with torch.no_grad():
    #     pred = model(tile_torch.unsqueeze(0))  # (1,C,p,p) or (1,num_classes,p,p)
    # pred_np = pred.squeeze(0).permute(1, 2, 0).cpu().numpy()
    
img_rec = merger.merge()
print(img_rec.shape)
print("Reconstruction MSE:", ((img - img_rec) ** 2).mean())
```

```python
masks_bin = {}
masks_seg = {}

patch_size = 256
overlap = 64 # overlap to avoid edge artifacts
batch_size = 8
img_shape = img.shape # run the previous cell first
C, H, W = img_shape

tiler = Tiler(
    data_shape=img_shape,
    tile_shape=(C, patch_size, patch_size),
    overlap=overlap,
    channel_dimension=0
)
merger = Merger(tiler)
binary_merger = Merger(Tiler(data_shape=img_shape,tile_shape=(2,patch_size,patch_size),overlap=overlap,channel_dimension=0))
nuclei_merger = Merger(Tiler(data_shape=img_shape,tile_shape=(10,patch_size,patch_size),overlap=overlap,channel_dimension=0))

for tid in tissue_ids[:1]:
    img = get_image(tid).numpy()  # C,H,W
    result_masks = []
    
    for batch_id, batch in tiler(img, batch_size=batch_size):
        # batch is (B, C, P, P)
        with torch.no_grad():
            with torch.autocast(device_type="cuda", dtype=torch.float16):
                batch_gpu = torch.from_numpy(batch).to(device)
                # dict_keys(['tissue_types', 'nuclei_binary_map', 'hv_map', 'nuclei_type_map'])
                predictions = model(batch_gpu, retrieve_tokens=True)
        result_masks.append(predictions['nuclei_binary_map'].cpu().numpy())
        binary_merger.add_batch(batch_id, batch_size, predictions['nuclei_binary_map'].cpu().numpy())
        nuclei_merger.add_batch(batch_id, batch_size, predictions['nuclei_type_map'].cpu().numpy())

    img_bin = binary_merger.merge()
    img_seg = nuclei_merger.merge()
    masks_bin[tid] = img_bin # numpy (2, H, W)
    masks_seg[tid] = img_seg
```

```python
values, counts = np.unique(masks_bin['orion_zupwxfxb_0'].argmax(axis=0), return_counts=True)
print(dict(zip(values, counts)))

masks_vis = [result_masks[0][0].argmax(axis=0), masks_bin['orion_zupwxfxb_0'].argmax(axis=0)]
titles = ["First Patch Prediction (Artifact Bottom Edge)", "Whole Image Prediction"]
plt.figure(figsize=(10, 5))
for ind, mask, title in zip(range(len(masks_vis)), masks_vis, titles):
    plt.subplot(1, 2, ind + 1)
    plt.imshow(np.ma.masked_where(mask == 0, mask), cmap='tab10', alpha=0.9, vmin=0, vmax=9)
    plt.title(title)
plt.tight_layout()
```

```python
from sklearn.metrics import f1_score, confusion_matrix, classification_report, accuracy_score

for tid, mask_pred in masks.items():
    mask_true_flat = get_mask(tid).flatten()
    mask_pred_flat = mask_pred.argmax(axis=0).flatten()

    accuracy = accuracy_score(mask_true_flat, mask_pred_flat)
    f1 = f1_score(mask_true_flat, mask_pred_flat, average='weighted')
    print(f"Tissue ID: {tid}, Accuracy: {accuracy:.4f}, F1 Score: {f1:.4f}")
```
