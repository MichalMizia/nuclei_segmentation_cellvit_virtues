# CellViT Inference Code

This notebook is separated into 2 primary stages:

1. Exploratory data analysis - Visualization of H&E images and their masks, showing different cell types etc.
2. Binary nuclei segmentation using CellVitSam model.

Semantic segmentation is not possible using only the pretrained CellVit models. Their models have different classification taxonomies, which can be found at: https://tio-ikim.github.io/CellViT-Inference/examples.html#classification-taxonomies

Their different cell types, cause the nuclei_type_map decoder to classify cells into different categories than our broad_cell_type

For that issue we have 2 possible solutions which are explored later:

- Mapping our broad_cell_type (10 classes) to one of the classification taxonomies. However benchmarking this classification agains a decoder fine-tuned on VirTues for 10 classes would be unfair as prediction tasks on aggregated cell types might be found easier for the model.
- Training the CellVit decoder again for predicting our broad_cell_type, while keeping the encoder frozen, this seems like the most fair option as we could train the decoder for both CellVit and Virtues encoder, using the same epochs, optimizer, and parameters to benchmark the representation quality.

````python
CELL_TYPE_LABELS = [
    None, "Stroma", "Endothelia", "Macrophage", "Tumor",
    "Helper T cell", "Cytotoxic T cell", "Other", "Treg", "B cells"
] # labels from the label encoder

def visualize_images(ds, num_cols=1):
    """Visualize two rows of tissue images:
    - First row: raw tissue images
    - Second row: tissue images overlaid with cell type masks
    Args:
        ds: multimodal dataset
        num_cols: number of tissue images to visualize
    """
    he_ds = ds.unimodal_datasets['he']
    tissue_ids = ds.get_tissue_ids()
    cmap = plt.get_cmap('tab10')

    plt.figure(figsize=(num_cols * 5, 2 * 5))
    plt.suptitle("Tissue Images")
    for i in range(num_cols):
        img = rearrange(he_ds._get_tissue_all_channels(tissue_ids[i]), 'c h w -> h w c')
        img = (img - img.min()) / (img.max() - img.min())
        mask = ds.get_cell_mask(tissue_ids[i], task="broad_cell_type", resize=True)
        
        plt.subplot(2, num_cols, i + 1)
        plt.imshow(img)
        plt.axis('off')
        plt.title(f"Tissue {i+1}")

        # Second row: image + mask
        plt.subplot(2, num_cols, num_cols + i + 1)
        plt.imshow(img)
        plt.imshow(np.ma.masked_where(mask == 0, mask), cmap=cmap, alpha=0.9, vmin=0, vmax=9)
        plt.axis('off')
        plt.title(f"Tissue {i+1} + Mask")

        if i == 0:
            unique_vals = np.unique(mask)
            legend_elements = []
            for val in unique_vals:
                if val == 0 or CELL_TYPE_LABELS[val] is None:
                    continue
                legend_elements.append(
                    Patch(facecolor=cmap(val), edgecolor='k', label=CELL_TYPE_LABELS[val])
                )
            if legend_elements:
                plt.legend(handles=legend_elements)
````

### Tiler library usage

Tiler library is used for splitting the image into 256x256 patches that work with the CellVit encoder. Overlap is needed because cell segmentation often produces artifacts on image boundaries. Merger object is used and with the overlap=64px, it merges last 32px from previous patch and first 32px from next patch. This solution avoids taking any pixels which would be close enough to image boundaries to cause artifacts

````python
# function from cellvit/inference/inference.py
def apply_softmax_reorder(predictions: dict) -> dict:
    """Reorder and apply softmax on predictions"""
    predictions["nuclei_binary_map"] = F.softmax(predictions["nuclei_binary_map"], dim=1)
    predictions["nuclei_type_map"] = F.softmax(predictions["nuclei_type_map"], dim=1)
    predictions["nuclei_type_map"] = predictions["nuclei_type_map"].permute(0, 2, 3, 1)
    predictions["nuclei_binary_map"] = predictions["nuclei_binary_map"].permute(0, 2, 3, 1)
    predictions["hv_map"] = predictions["hv_map"].permute(0, 2, 3, 1)
    return predictions
````

````python
masks_bin = {}
masks_seg = {}

patch_size = 256
overlap = 64 # overlap to avoid edge artifacts
batch_size = 8

img = get_image(tissue_ids[0]).numpy() # sample image for shape
img_shape = img.shape # run the previous cell first
C, H, W = img_shape

tiler = Tiler(
    data_shape=img_shape,
    tile_shape=(C, patch_size, patch_size),
    overlap=overlap,
    channel_dimension=0
)
binary_merger = Merger(Tiler(data_shape=img_shape,tile_shape=(2,patch_size,patch_size),overlap=overlap,channel_dimension=0))
nuclei_merger = Merger(Tiler(data_shape=img_shape,tile_shape=(10,patch_size,patch_size),overlap=overlap,channel_dimension=0))
````

````python
for tid in tqdm(tissue_ids):
    img = get_image(tid).numpy()  # H, W, C
    result_masks = []
    
    for batch_id, batch in tiler(img, batch_size=batch_size):
        # batch is (B, C, P, P)
        # print(batch_id, batch.shape)
        with torch.no_grad():
            with torch.autocast(device_type="cuda", dtype=torch.float16):
                batch_gpu = torch.from_numpy(batch).to(device)
                # dict_keys(['tissue_types', 'nuclei_binary_map', 'hv_map', 'nuclei_type_map'])
                predictions = model(batch_gpu, retrieve_tokens=True)
        result_masks.append(predictions['nuclei_binary_map'].cpu().numpy())
        binary_merger.add_batch(batch_id, batch_size, predictions['nuclei_binary_map'].cpu().numpy())
        nuclei_merger.add_batch(batch_id, batch_size, predictions['nuclei_type_map'].cpu().numpy())
        
    img_bin = binary_merger.merge()
    img_seg = nuclei_merger.merge()
    masks_bin[tid] = img_bin # numpy (2, H, W)
    masks_seg[tid] = img_seg # numpy (10, H, W)
````

````python
values, counts = np.unique(masks_bin['orion_zupwxfxb_0'].argmax(axis=0), return_counts=True)
print(f"Value counts for whole image: {dict(zip(values, counts))}")

plot_masks = [result_masks[0][0].argmax(axis=0), masks_bin['orion_zupwxfxb_0'].argmax(axis=0)]
titles = ["First Patch Prediction (Artifacts)", "Whole Image Prediction (orion_zupwxfxb_0)"]
plt.figure(figsize=(12, 5))
for ind, mask, title in zip(range(len(plot_masks)), plot_masks, titles):
    plt.subplot(1, 2, ind + 1)
    plt.imshow(np.ma.masked_where(mask == 0, mask), cmap='tab10', alpha=0.9, vmin=0, vmax=9)
    plt.title(title)
plt.tight_layout()
````

````python
from sklearn.metrics import f1_score, confusion_matrix, classification_report, accuracy_score
# we do a 80/20 train/test split
# From starting 35 tissue images, we have 28 train images and 7 test images
# We measure the accuracy only on test images to benchmark later against tuned models
train_len = int(0.8 * len(tissue_ids))
train_tids = tissue_ids[:train_len]
test_tids = tissue_ids[train_len:]

accuracies, f1s = [], []
for tid, mask_pred in masks_bin.items():
    if tid in train_tids:
        continue
        
    mask_true_flat = get_mask(tid).flatten()
    mask_pred_flat = mask_pred.argmax(axis=0).flatten()

    accuracy = accuracy_score(mask_true_flat, mask_pred_flat)
    f1 = f1_score(mask_true_flat, mask_pred_flat, average='weighted')
    print(f"Tissue ID: {tid}, Accuracy: {accuracy:.4f}, F1 Score: {f1:.4f}")

    accuracies.append(accuracy)
    f1s.append(f1)

accuracies = np.array(accuracies)
f1s = np.array(f1s)

print(f"Mean Accuracy: {accuracies.mean():.4f}, Mean F1 Score: {f1s.mean():.4f}")
````
